# Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('articles.csv')
pd.set_option('display.max_columns',100)
df
```

```python
df.dtypes
```

```python
df.isna().any()
```

```python
df[df['detail_desc'].isna()].shape
```

```python
df_cat = df.loc[:,'product_group_name'::2]
df_cat['article_id'] = df['article_id']
df_cat['index_code'] = df['index_code']
df_cat['product_type_name'] = df['product_type_name']
df_cat['prod_name'] = df['prod_name']
df_cat
```

```python
df_cat.dtypes
```

```python
df_num = df.loc[:,'graphical_appearance_no'::2]
df_num.drop(axis=1, inplace=True, columns=['index_code','detail_desc'])
df_num['article_id'] = df['article_id']
df_num['product_type_no'] = df['product_type_no']
df_num['product_code'] = df['product_code']
df_num
```

```python
df_num.dtypes
```

```python
import json
  
# Opening JSON file
f = open('articles.json')
  
# returns JSON object as a dictionary
articles = json.load(f)
articles
```

```python
len(articles)
```

```python
df_cat['article_id'] = df_cat['article_id'].apply(lambda x: articles[str(x)] if str(x) in articles.keys() else x)
df_cat.sort_values('article_id')
```

```python
df_cat = df_cat[df_cat['article_id']<50000]
df_cat
```

```python
df_cat.to_csv('articles_cat.csv',index=False)
```

```python
df_num['article_id'] = df_num['article_id'].apply(lambda x: articles[str(x)] if str(x) in articles.keys() else x)
df_num.sort_values('article_id')
```

```python
df_num = df_num[df_num['article_id']<50000]
df_num
```

```python
df_num.to_csv('articles_num.csv',index=False)
```
