```python
#df_c = df.groupby('customer_id').count().sort_values(by='t_dat')
df_c = df.groupby('customer_id',as_index=False).count()
df_c.rename(columns={"t_dat": "count"},inplace=True)
#df_c = df_c[['count']]
df_c
```

```python
df_c.describe()
```

1371980 - 1362281 = 9699 customers that have never purchased anything

```python
df_merged = df.merge(df_c,on = 'customer_id')
```

```python
df_merged
```

```python
df_cropped_c = df_merged[df_merged['count'] > 26]
df_cropped_c
```

```python
df_cropped_c[df_cropped_c.duplicated(subset=['customer_id','article_id_x'])]
```

```python
df_cropped_c = df_cropped_c.loc[:,'t_dat':'sales_channel_id_x']
```

```python
#df_a = df.groupby('article_id').count().sort_values(by='t_dat')
df_a = df.groupby('article_id',as_index=False).count()
df_a.rename(columns={"t_dat": "count"},inplace=True)
#df_a = df_a[['count']]
df_a
```

```python
df_a.describe()
```

105542 - 104547 = 995 products that haven't been purchased

```python
df_merged_2 = df_cropped_c.merge(df_a,right_on = 'article_id',left_on='article_id_x')
df_merged_2
```

```python
df_f = df_merged_2[df_merged_2['count']>100]
```

```python
df_f
```

```python
df_f = df_f.loc[:,'t_dat':'sales_channel_id_x']
```

```python
df_f=df_f.rename(columns={'customer_id_x':'user_id','article_id_x':'item_id','price_x':'price','sales_channel_id_x':'sales_channel_id'})
```

# Whole dataset

```python
df_f = df
```

# Mapping

```python
articles = {}
customers = {}
```

```python
def map_artcl(s):
    return articles[s]

def map_cstmr(s):
    return customers[s]
```

```python
c_id = 0
for index,row in df_f.iterrows():
    if not row['customer_id'] in customers.keys():
        customers[row['customer_id']] = c_id
        c_id += 1
customers
```

```python
c_id
```

```python
a_id = 0
for index,row in df_f.iterrows():
    if not row['article_id'] in articles.keys():
        articles[row['article_id']] = a_id
        a_id += 1
articles
```

```python
a_id
```

```python
df_f['customer_id'] = df_f['customer_id'].apply(map_cstmr)
```

```python
df_f['article_id'] = df_f['article_id'].apply(map_artcl)
```
