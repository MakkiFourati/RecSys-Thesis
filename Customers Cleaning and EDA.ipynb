# Further pre processing

```python
#df_e = pd.read_csv('customers_cleaned.csv')
df_e = df_f
```

```python
df_e
```

```python
df_e.groupby('postal_code').count().describe()
```

```python
df_ee = df_e.copy()
```

```python
df_ee['postal_code'] = df_e['postal_code'].apply(lambda s: s[:-5:-1])
```

```python
df_ee
```

```python
df_ee.groupby('postal_code').count().describe()
```

```python
df_e = df_e.sort_values('postal_code')
```

```python
df_e
```

```python
postal_codes = {}
i = 0
for index, row in df_e.iterrows():
    if not row['postal_code'] in postal_codes.keys():
        postal_codes[row['postal_code']] = i
        i+=1
```

```python
postal_codes
```

```python
len(postal_codes)
```

```python
df_e['postal_code'] = df_e['postal_code'].apply(lambda x: postal_codes[x])
df_e
```

```python
df_e.dtypes
```

```python
df_e.to_csv('customers_cleaned_1.csv', index=False)
```
